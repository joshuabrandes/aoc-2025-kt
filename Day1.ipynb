```kotlin
val input = aocClient.input()
val start = 50
// this is a rotating dial lock going from 0 to 99

fun next(current: Int, command: String): Int {
    // command is like L21 or R88; L is substraction R is addition
    val dir = command.first()
    val amount = command.drop(1).toInt()

    fun calculate(curr: Int, change: Int): Int {
        val temp = curr + change
        //ensure that it is between 0 and 99
        return temp % 100
    }

    return when (dir) {
        'L' -> calculate(current, -amount)
        'R' -> calculate(current, amount)
        else -> throw IllegalArgumentException("Invalid direction: $dir")
    }
}

var lines = input.split("\n")
var values = mutableListOf(start)
for (line in lines) {
    values.add(next(values.last(), line))
}
val answerPt1 = values.filter { it == 1 }.count()
aocClient.submitPartOne(answerPt1)
```
