```kotlin
val input = aocClient.input()
val start = 50
// this is a rotating dial lock going from 0 to 99

fun next(current: Int, command: String): Int {
    // command is like L21 or R88; L is substraction R is addition
    val dir = command.first()
    val amount = command.drop(1).toInt()

    fun calculate(curr: Int, change: Int): Int {
        var temp = curr + change
        //ensure that it is between 0 and 99
        if (temp < 0) temp += 100
        return temp % 100
    }

    return when (dir) {
        'L' -> calculate(current, -amount)
        'R' -> calculate(current, amount)
        else -> throw IllegalArgumentException("Invalid direction: $dir")
    }
}

var lines = input.split("\n")
var values = mutableListOf(start)
for (line in lines) {
    values.add(next(values.last(), line))
}
val answerPt1 = values.filter { it == 0 }.count()
aocClient.submitPartOne(answerPt1)
```

```kotlin
val lines2 = input.split("\n")
fun countZerosDuringRotations(start: Int, commands: String): Pair<Int, Int> {
    val dir = commands.first()
    val amount = commands.drop(1).toInt()

    var current = start
    var zeroCount = 0

    val step = if (dir == 'L') -1 else 1

    repeat(amount) {
        current = (current + step + 100) % 100
        if (current == 0) {
            zeroCount++
        }
    }

    return Pair(current, zeroCount)
}

var currentPosition = start
var totalZeros = 0
for (line in lines2) {
    val (newPosition, zeros) = countZerosDuringRotations(currentPosition, line)
    currentPosition = newPosition
    totalZeros += zeros
}

val answerPt2 = totalZeros
aocClient.submitPartTwo(answerPt2)
```
